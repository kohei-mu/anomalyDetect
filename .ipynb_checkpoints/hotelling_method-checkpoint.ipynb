```r
x1 <- rnorm(300)
x2 <- rep(c(15,rep(0,49)),6)
x3<- c(rep(0,150),rep(3,150)) 
x <- x1+x2+x3
```

```r
plot(x, type="l", main="異常値挿入前",ylab="", xlab="")
```

```r
y <- x+c(rep(0,130),30,rep(0,169))
plot(y, type="l", main="異常値挿入後", xlab="", ylab="")
```

```r
plot(density(y), main="極少数の異常データ", ylab="", xlab="")
```

```r
mu <- mean(y)
s2 <- mean((y - mu)^2)
```

```r
anom <- (y - mu)^2 / s2
```

```r
thr <- qchisq(0.99, 1)
```

```r
plot(anom, type="l", main="ホテリングのT２乗 - 異常スコア", ylab="異常スコア", xlab="")
segments(0, thr, 300, thr, col='red', lty=3, lwd=3)
```

```r
plot(y, type="l", main="ホテリング理論により検知したデータ", xlab="", ylab="")
segments(0, thr, 300, thr, col='red', lty=3, lwd=3)
```
